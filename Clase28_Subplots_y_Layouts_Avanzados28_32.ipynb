```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Creamos valores función seno

# Generamos 100 valores entre 0-10
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creamos datos aleatorios
data = np.random.randn(100)
```

```python
# Generar cuadrícula 2x2
gs = gridspec.GridSpec(2, 2, height_ratios=[2,1], width_ratios=[1,1])
# Personalizamos figura
fig = plt.figure(figsize=(10,8))

# Primer subplot grande, ocupa toda la primera fila
ax1 = fig.add_subplot(gs[0,:])
# Generar y personalizar gráfico
ax1.plot(x,y, color='blue')
# Etiquetar
ax1.set_title('Seno de X')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

# Segundo subplot, ocupa la esquina inferior izquierda
ax2 = fig.add_subplot(gs[1,0])
# Generar y personalizar gráfico
ax2.hist(data, bins=20, color='purple', edgecolor='black')
# Etiquetar
ax2.set_title('Histograma')
ax2.set_xlabel('Valor')
ax2.set_ylabel('Frecuencia')

# Tercer subplot, ocupa la esquina inferior derecha
ax3 = fig.add_subplot(gs[1,1])
# Generar y personalizar gráfico
ax3.scatter(x,y, color='red')
# Etiquetar
ax3.set_title('Dispersión de seno')
ax3.set_xlabel('Valor')
ax3.set_ylabel('Frecuencia')

# Ajustar layout para evitar superposición de gráficas
plt.tight_layout()

plt.show()
```
