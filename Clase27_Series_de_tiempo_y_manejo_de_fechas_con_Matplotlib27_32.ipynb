```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
# Crear df de 100 fechas
dates = pd.date_range(start='2023-01-01', periods=100)

# Crear 100 valores pero acumulando la suma, es decir sumando los primeros números
values = np.random.rand(100).cumsum()

# Crear df con las fechas y valores generados
data = pd.DataFrame({'Date': dates, 'Value': values})
```

```python
# Crear gráfico de líneas mediate dos objeots fig, ax
fig, ax = plt.subplots(figsize=(12,6))

# Configurar gráfico
ax.plot(data['Date'], data['Value'], color='green')

# Configuramos etiquetas
plt.xticks(rotation=45) # etiquetas eje x a 45°

plt.title('Serie de tiempo con formato en las fechas')
plt.xlabel('Date')
plt.ylabel('Value')

plt.show()
```

```python
# Reto visualizar las ventas mensuales

# Crear df con 12 fechas desde enero hasta diciembre 2023, frequency = M por mes
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
# Crear 12 valores (ventas)
sales = np.random.randint(1000, 5000, size=12)

# Creamos df con las 12 fechas y las 12 valores (ventas)
sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

# Crear gráfico
plt.plot(sales_data['Date'], sales_data['Sales'], marker='o', linestyle='-', label='Ventas mensuales')

# Etiquetar gráfico
plt.xticks(rotation=45)
plt.title('Análisis de ventas mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()

plt.tight_layout()

plt.show()
```

```python
# Personalizamos foramto de fecha en gráfico

# Reto visualizar las ventas mensuales

# Crear df con 12 fechas desde enero hasta diciembre 2023, frequency = M por mes
dates = pd.date_range(start='2023-01-01', periods=12, freq='M')
# Crear 12 valores (ventas)
sales = np.random.randint(1000, 5000, size=12)

# Creamos df con las 12 fechas y las 12 valores (ventas)
sales_data = pd.DataFrame({'Date': dates, 'Sales': sales})

# Crear gráfico
plt.plot(sales_data['Date'], sales_data['Sales'], marker='o', linestyle='-', label='Ventas mensuales')

# Formateamos fechas del gráfico creado
plt.gca().xaxis.set_major_formatter(DateFormatter('%b %Y'))

# Etiquetar gráfico
plt.xticks(rotation=45)
plt.title('Análisis de ventas mensuales')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.legend()

plt.tight_layout()

plt.show()
```
