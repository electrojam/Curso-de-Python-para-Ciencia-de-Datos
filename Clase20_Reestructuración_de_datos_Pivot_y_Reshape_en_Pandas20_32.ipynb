```python
import pandas as pd
```

```python
# Leer dataset de archivo csv
file_path = 'https://raw.githubusercontent.com/databricks/Spark-The-Definitive-Guide/refs/heads/master/data/retail-data/all/online-retail-dataset.csv'
sales_data = pd.read_csv(file_path)
```

```python
# Convertir columna 'InvoiceDate' a tipo datetime
sales_data['InvoiceDate'] = pd.to_datetime(sales_data['InvoiceDate'])
```

```python
# Eliminar filas con valores faltantes en columnas críticas 'CustomerID', 'InvoiceDate'
sales_data.dropna(subset=['Quantity', 'InvoiceDate'], inplace = True)
```

```python
# Crear columna 'TotalPrice' = 'Quantity' * 'UnitPrice'
sales_data['TotalPrice'] = sales_data['Quantity'] * sales_data['UnitPrice']
```

```python
sales_data.head()
```

```python
# pivot table
# sumar valores de Quantity, filas o index 'Country', columnas 'StockCode'
pivot_table = pd.pivot_table(sales_data,
                             values='Quantity',
                             index='Country',
                             columns='StockCode',
                             aggfunc='sum')
print(pivot_table)
```

```python
# Transformar df, apilar y desapilar información
df = pd.DataFrame({
    'A': ['foo', 'bar','baz'],
    'B': [1, 2,3],
    'C': [4, 5, 6]
})

print(df)
```

```python
# Apilar stack
df_stack = df.stack()
print(df_stack)
```

```python
# Desapilar
df_unstack = df_stack.unstack()
print(df_unstack)
```
