```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/databricks/Spark-The-Definitive-Guide/refs/heads/master/data/retail-data/all/online-retail-dataset.csv')
print(df.head())
```

```python
df.info()
```

```python
# Convertimos 'InvoiceDate' de object a date
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

```python
# Eliminar valores nulos, modificar df en sitio, no traer copia.
df.dropna(subset=['InvoiceDate'], inplace=True)
```

```python
# Poner como index la columna 'InvoiceDate', modificar df en sitio, no traer copia.
df.set_index('InvoiceDate', inplace = True)
```

```python
# Crear columna 'Year', llenar con año extraiDo de index 'InvoiceDate'
df['Year'] = df.index.year
df['Month'] = df.index.month
df['Day'] = df.index.day
df['Weekday'] = df.index.weekday
df['Hour'] = df.index.hour

print(df['Year'])
print(df['Hour'])
```

```python
df
```

```python
# Extraer con .loc filas con index 2011 ( 2011)
df_2011 = df.loc['2011']
df_2011.head()
```

```python
# Extraer con .loc filas con index 2011-12 (dic 2011)
df_2011_dec = df.loc['2011-12']
df_2011_dec.head()
```

```python
# Extraer con loc rango de fecha de index 2010-12-01 hasta 2010-12-15
df_dec_range = df.loc['2010-12-01':'2010-12-15']
df_dec_range
```

```python
# Crear rangos de fechas
date_range_new = pd.date_range(start='2024-01-01', end='2024-12-31', freq='D')
print(date_range_new)
```

```python
# Crear df serie con date_range_new, nombre de columna 'Date
df_dates = pd.DataFrame(date_range_new, columns=['Date'])
df_dates
```
