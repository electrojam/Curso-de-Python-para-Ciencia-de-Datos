```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creamos array con datos aleatorios con distribuci칩n normal
# Datos de alturas de personas
# con una media = 170, std = 10 , 200 datos.

data = np.random.normal(170, 10, 200)
print(data)
```

```python
# Graficamos histrogram de data

# Creamos el gr치fico histograma
plt.hist(data, color='salmon', edgecolor='black', alpha=0.6)

# Etiquetamos
plt.title('Distribuci칩n de alturas de personas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')

# Mostramos el gr치fico
plt.show()
```

```python
# Modificamos divisiones, n칰mero de intervalos del gr치fico histograma

# Creamos el gr치fico histograma
plt.hist(data, color='salmon', edgecolor='black', alpha=0.6, bins=20)

# Etiquetamos
plt.title('Distribuci칩n de alturas de personas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')

# Mostramos el gr치fico
plt.show()
```

```python
# Gr치fico de Boxplot

# Extablecer una semilla de inicio
# para que no cambie la data cada que se ejecute
np.random.seed(0)

# Crear lista con 3 arrays con informaci칩n de rango de edades
ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]

print(ages)
```

```python
# Creamos gr치fico Boxplot con los 3 datas
plt.boxplot(ages, patch_artist=True, notch=True, vert=True, labels=['Grupo', 'Grupo 2', 'Grupo 3'])

# Etiquetamos el gr치fico
plt.title('Distribuci칩n de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (a침os)')

# Mostramos el gr치fico
plt.show()
```
